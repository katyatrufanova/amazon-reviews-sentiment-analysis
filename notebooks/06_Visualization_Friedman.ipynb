```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
models = ['Naive Bayes', 'Logistic Regression', 'SVM', 'biLSTM with GloVe', 'biLSTM without GloVe', 'DistilBERT']
accuracy = [0.85, 0.87, 0.87, 0.92, 0.92, 0.93]
precision = [0.70, 0.76, 0.77, 0.85, 0.85, 0.86]
recall = [0.60, 0.67, 0.65, 0.81, 0.82, 0.84]
f1_score = [0.61, 0.68, 0.66, 0.82, 0.83, 0.85]
```

```python

N = len(models)

ind = np.arange(N)
plt.figure(figsize=(10,5))
width = 0.2

plt.bar(ind, accuracy, width, label='Accuracy', color='#e6194B')
plt.bar(ind + width, precision, width, label='Precision', color='#3cb44b')
plt.bar(ind + width * 2, recall, width, label='Recall', color='#ffe119')
plt.bar(ind + width * 3, f1_score, width, label='F1-score', color='#4363d8')

plt.xlabel('Models')
plt.ylabel('Scores')

plt.xticks(ind + width / 2, models, rotation=30)

plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
from scipy.stats import friedmanchisquare
import numpy as np
```

```python
naive_bayes_scores = np.array([0.84323476, 0.84375818, 0.84450593, 0.84304782, 0.84230007, 0.8460762, 0.8451789, 0.8401316, 0.84569655, 0.84397667])
logistic_regression_scores = np.array([0.86895727, 0.87015366, 0.87037799, 0.87161177, 0.87007889, 0.86910682, 0.87277078, 0.86921898, 0.87104614, 0.87254169])
svm_scores = np.array([0.86727483, 0.86783565, 0.86839646, 0.86936853, 0.86813474, 0.86813474, 0.87105096, 0.86746177, 0.87059747, 0.87059747])

stat, p = friedmanchisquare(naive_bayes_scores, logistic_regression_scores, svm_scores)

print('F statistic:', stat)
print('p-value:', p)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
data = pd.DataFrame({
    'Naive Bayes': naive_bayes_scores,
    'Logistic Regression': logistic_regression_scores,
    'SVM': svm_scores
})
```
